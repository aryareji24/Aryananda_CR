```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
path = "C:\\Users\\winsa\\B_Pandas\\main.xlsx"
df = pd.read_excel(path)
```

```python
def matplot_lib_grouped(df, x, y1, y2, y3):
    df.plot(x=x, y=[y1, y2, y3], kind='bar')
    plt.xlabel(x)
    plt.ylabel('Values')
    plt.title(' Matplotlib : Grouped Bar Chart')
    plt.show()
```

```python
def seaborn_grouped(df, x, y1, y2, y3):
    df_melted = pd.melt(df, id_vars=x, value_vars=[y1, y2, y3], var_name='Category', value_name='Value')
    sns.barplot(data=df_melted, x=x, y='Value', hue='Category')
    plt.xlabel(x)
    plt.ylabel('Values')
    plt.title('Seaborn : Grouped Bar Chart')
    plt.show()
```

```python
def plotly_grouped(df, x, y1, y2, y3):
    df_melted = df.melt(id_vars=x, value_vars=[y1, y2, y3], var_name='Category', value_name='Value')
    chat = px.bar(df_melted, x=x, y='Value', color='Category', width=800,
              height=600,barmode='group', title='Plotly : Grouped Bar Chart ')
    chat.show()
```

```python
df['report_date_time']= pd.to_datetime(df['report_date_time'],format='%d-%m-%Y %H:%M')
df_yearly = df.groupby(['Year'])[['me_con', 'ae_con', 'bl_con']].sum().reset_index()
```

```python
matplot_lib_grouped(df_yearly, 'Year', 'me_con', 'ae_con', 'bl_con')
seaborn_grouped(df_yearly, 'Year', 'me_con', 'ae_con', 'bl_con')
plotly_grouped(df_yearly, 'Year', 'me_con', 'ae_con', 'bl_con')
```
