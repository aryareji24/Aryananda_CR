### BubbleChart Using matplotlib

```python

import matplotlib.pyplot as plt

df['report_date_time'] = pd.to_datetime(df['report_date_time'])

plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    df['report_date_time'], 
    df['Final_Total_FO'], 
    s=df['total_fo'] * 10,  # Scale up the size of bubbles
    c=df['total_fo'],  # Color by 'total_fo'
    cmap='viridis',  # Use a colormap
    alpha=0.6, 
    edgecolor='w',  # White edge color for bubbles
    linewidth=0.5
)

# Adding a colorbar
plt.colorbar(scatter, label='Total FO')

plt.xlabel('Date Time')
plt.ylabel('Final Total FO')
plt.title('Bubble Chart of Final Total FO by Date Time with Colormap')

plt.show()
```
