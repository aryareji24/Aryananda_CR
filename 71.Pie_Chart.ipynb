### Pie Chart Using matplotlib

```python
import matplotlib.pyplot as plt

# Group the data by Month_Name and sum the Final_Total_FO for each month
month_data = df.groupby('Month_Name')['Final_Total_FO'].sum()

plt.figure(figsize=(8, 8))
plt.pie(
    month_data, 
    labels=month_data.index, 
    autopct='%1.1f%%',  # Display percentages
    startangle=140  # Rotate the chart for a better visual angle
)

plt.title('Distribution of Final Total FO by Month')

plt.show()
```
