# RANDOM MODULE

## The Python random module is a built-in module that provides functionalities related to random number generation.

### * seed() : Initialize the random number generator

```python
import random

random.seed(10)
print(random.random())
```

* The seed() method is used to initialize the random number generator.

* The random number generator needs a number to start with (a seed value), to be able to generate a random number.

### * getstate()	Returns the current internal state of the random number generator

```python
import random

print(random.getstate())
```

* The getstate() method returns an object with the current state of the random number generator.

### * setstate()	Restores the internal state of the random number generator

```python
import random
print(random.random())
state = random.getstate()

print(random.random())
random.setstate(state)

print(random.random())
```

### * random.choice(sequence) - Returns a random element from the given sequence

```python
import random
l=[1,2,3,4,5,6,7]
print(random.choice(l)) 
```

 ### * random.random() - Returns a random float between 0 and 1

```python
import random
random.seed(5)
srandom =[random.random() for _ in range(3)]
print(srandom)
```

### *  random.randint(a, b) - Returns a random integer between a given range a and b i.e. inclusive i.e. a=start, b

```python
r1 = random.randint(15,25)
print("Random number between 15 and 25 is",r1)
r2 = random.randint(-11, -2)
print("Random number between -11 and -2 is",r2)
```

### * random.uniform(a, b) - Returns a random float between a and b

```python
prices = [random.uniform(10, 40) for _ in range(5)]
print(prices)
```

### * random.sample(items, k) - Returns a k-length list of unique elements chosen from the items

```python
products = ['Laptop', 'Phone', 'Tablet', 'Mouse', 'Camera']
selected = random.sample(products, 3)
print(selected)
```
